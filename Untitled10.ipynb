# Room types

We also look up the ‘room_type’ columns for each room type of the listing

```python
df['room_type'].unique()
```

From the list above, we see that Airbnb have 4 room type. Based on the information on the Airbnb website, the definition of each room type are:

# Private room
Guests have exclusive access to the bedroom/sleeping area of the listing. Other parts area such as the living room, kitchen, and bathroom are likely open either to the host even to other guests.
# Entire home/apt
Guests have the whole place for themselves. It usually includes a bedroom, bathroom, and kitchen.
# Shared Room
Guest sleep in a bedroom or a common area that could be shared with others.
# Hotel Room
A typical hotel room with its facilities. Since 2018, Airbnb allows some boutique hotels and high rated independent hotel to list their rooms on their site.

# minimum nights

distribution of the minimum number of nights for properties in the dataset. 

```python
df['minimum_nights'].nunique(),df['minimum_nights'].min(), df['minimum_nights'].max()
```

```python
# Seaborn histogram with x-axis limit, binwidth of 1, and x-ticks for minimum nights
plt.figure(figsize=(10, 6))
sns.histplot(df[df['minimum_nights'] <= 30], x='minimum_nights', binwidth=1, kde=False, color='blue')
plt.title('Distribution of Minimum Nights')
plt.xlabel('Minimum Nights')
plt.ylabel('Frequency')
plt.xticks(np.arange(0, 31, 1))
plt.show()
```

The majority of properties require a low minimum number of nights, typically between 1 to 5 nights. This is evident from the high bars on the left side of the histogram. This suggests that many hosts are open to short stays, which could include weekend getaways or short business trips.

As we move to the right along the x-axis, representing an increase in the minimum number of nights required, the frequency of properties requiring those longer minimum stays decreases. This is indicated by the decreasing height of the bars. This suggests that fewer properties are available for longer-term rentals, or that fewer hosts require longer minimum stays.

There are a few properties that require a high minimum number of nights, as indicated by the bars on the far right of the histogram. However, these properties are relatively rare compared to those requiring shorter stays.

Overall, the histogram shows a right-skewed distribution, indicating that most properties require a low to moderate minimum number of nights, with fewer properties requiring longer minimum stays.
